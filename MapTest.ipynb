```python
import osmnx as ox
import matplotlib.pyplot as plt
```

```python
# Graph from address
def graph_from_address_wrapper(address, distance=1000, distance_type='bbox', network_type='drive', fig_height=5, node_size=0):
    g = ox.graph_from_address(
        address=address, 
        distance=distance, 
        distance_type=distance_type, 
        network_type=network_type)
    
    project_and_plot(g, fig_height, node_size)
    
def graph_from_point_wrapper(point, distance=1000, distance_type='bbox', network_type='drive', fig_height=5, node_size=0):
    g = ox.graph_from_point(
        point, 
        distance=distance, 
        distance_type=distance_type, 
        network_type=network_type)
    
    project_and_plot(g, fig_height, node_size, point=point)

def project_and_plot(g, fig_height, node_size, point=None):
    # you can project the network to UTM (zone calculated automatically)
    proj = ox.project_graph(g)
       
    if point: # Don't show plot until point is added
        fig, ax = ox.plot_graph(proj, fig_height=fig_height, node_size=node_size, show=False, close=False)
        # I think this is the problem line below, I don't know why the red point isn't showing up ðŸ˜ 
        ax = ax.scatter(x=point[1], y=point[0], c='red')
        plt.show()
    else:
        fig, ax = ox.plot_graph(proj, fig_height=fig_height, node_size=node_size)
```
